### Data Visualization with Searborn

Seaborn is a Python Visualization Library based on Matplotlib that provides a high level interface for drawing attractive and informative statistical graphics, Seaborn helps in creating complex visualizations with just a few lines of code.

```python
import seaborn as sns
```

```python
# Basic plotting with seaborn
tips = sns.load_dataset('tips')
tips
```

```python
# Create a scatter plot
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill', y= 'tip', data=tips)
plt.title("Scatter Plot of total bill.")
plt.show()
```

```python
# Line plot
sns.lineplot(x='size', y='total_bill', data=tips)
plt.title("Time plot of total Bill by Size")
```

```python
# Categorical plots
# Bar plot
sns.barplot(x='day', y='total_bill', data=tips)
plt.title("Bar plot of total bill by day")
plt.show()
```

```python
# Box plot
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
# Violin plot
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
# Histogram
sns.histplot(tips["total_bill"], bins=10, kde=False) # Kernal density estimator
```

```python
sns.histplot(tips["total_bill"], bins=10, kde=True) # Kernal density estimator
```

```python
# KDE Plot
sns.kdeplot(tips['total_bill'], shade=True)
```

```python
sns.kdeplot(tips['total_bill'], fill=True) # Gives no warning
```

```python
# Pairplot
sns.pairplot(tips)
```

```python
# Heatmap
corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr)
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
```
